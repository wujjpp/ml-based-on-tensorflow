## 2. TensorFlow

<img src="./images/tensorflow.svg" width="200"/> 

2015年11月9日发行beta版本，TensorFlow支持多种客户端语言下的安装和运行。当前最新版本1.14.0， 2.0目前处于beta版状态。
绑定完成的语言有Python、JavaScript、C++、Go和Java，移动端 TensorFlow Lite、TensorFlow Lite on iOS。

> TensorFlow主要支持静态计算图的形式，计算图的结构比较直观，但是在调试过程中十分复杂与麻烦，一些错误更加难以发。但是2017年底发布了动态图机制Eager Execution，加入对于动态计算图的支持，但是目前依旧采用原有的静态计算图形式为主。 2.0版本默认启用Eager Execution

分：CPU版本和GPU版, 硬件可选: Nvidia Tesla显卡
